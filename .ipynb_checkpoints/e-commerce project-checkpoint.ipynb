```python
import pandas as pd 
import plotly.express as px # plotly advanced than matplotlib and seaborn and is used for data visualization
import plotly.graph_objects as go #for advanced and customized graphs
import plotly.io as pio #for graph templates
import plotly.colors as colors #for colors
pio.templates.default = "plotly_white"
```

```python
data=pd.read_csv("superstore_data.csv",encoding='latin-1')
```

```python
data
```

```python
data.head() #for first 5 rows 
```

```python
data.describe() #for descriptive stats only for numericals
```

```python
data.info() #for more information
```

# CONVERTING DATE COLUMNS


```python
data['Order Date']=pd.to_datetime(data['Order Date'], errors='coerce')
```

```python
data.info()
```

```python
data['Ship Date']=pd.to_datetime(data['Ship Date'], errors='coerce')
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of Week']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly Sales Analysis

```python
Sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig=px.line(Sales_by_month,
            x='Order Month',
            y='Sales',
            title='Monthly Sales Analysis')
fig.show()
```

# Lowest monthly sales - 1(january) and highest monthly sales - 11 (november)

```python
Sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
Sales_by_month
```

# Sales analysis by category

```python
data.head()
```

```python
Sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
Sales_by_category
```

```python
Sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
fig=px.line(Sales_by_category,
            x='Category',
            y='Sales',
            title='CATEGORY WISE SALES ANALYSIS')
fig.show()
```

# Lowest sales of category - Office Supplies and highest sales of category - Technology

```python
fig= px.pie(Sales_by_category,
            values='Sales',
            names='Category',
            hole=0.5,
            color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category' , title_font=dict(size=24) )

fig.show()
```

# Sales analysis by sub-category


```python
data.head()
```

```python
Sales_by_subcategory=data.groupby('Sub-Category')['Sales'].sum().reset_index()
Sales_by_subcategory
```

```python
fig=px.line(Sales_by_subcategory,
            x='Sub-Category',
            y='Sales',
            title='Sales analysis by subcategory')
fig.show()
```

```python
fig=px.bar(Sales_by_subcategory,
           x='Sub-Category',
           y='Sales',
           title='Sales analysis by subcategory using bar graph')
fig.show()
```

# highest sales = chairs and lowest sales = fasteners

# Sales and profit - customer segment

```python
Sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_palette=colors.qualitative.Pastel

fig=go.Figure()
fig.add_trace(go.Bar(x = Sales_profit_by_segment['Segment'],
                     y= Sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=Sales_profit_by_segment['Segment'],
                     y=Sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment',
                  yaxis_title='Amount')
fig.show()
```

# Sales to Profit ratio

```python
Sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
Sales_profit_by_segment['Sales_to_Profit_ratio']=Sales_profit_by_segment['Sales']/Sales_profit_by_segment['Profit']
print(Sales_profit_by_segment[['Segment','Sales_to_Profit_ratio']])
```

```python
data.info()
```

# location and time period analysis

```python
top_15_states = data['State'].value_counts().nlargest(15).index
heatmap_data = data[data['State'].isin(top_15_states)]

matrix_data = heatmap_data.groupby(['Order Month', 'State']).size().reset_index(name='Order_Count')

fig = px.density_heatmap(
    matrix_data, 
    x="Order Month", 
    y="State", 
    z="Order_Count",
    color_continuous_scale="Viridis",
    title="Heatmap: Order Volume by State & Time Period",
    labels={'Order Month': 'Timeline', 'Order_Count': 'Number of Orders'},
    category_orders={"State": list(top_15_states)} # Keeps most active states at the top
)


fig.update_layout(xaxis_nticks=20) 
fig.show()
```

```python
data['Ship Duration'] = (data['Ship Date'] - data['Order Date']).dt.days
```

# Product category and shipping time analysis

```python
fig = px.box(
    data, 
    x="Category", 
    y="Ship Duration", 
    color="Category",
    points="all", 
    title="Shipping Duration by Product Category",
    labels={'Ship Duration': 'Days to Ship'},
    template="plotly_white"
)

fig.show()
```

```python
ship_counts = data.groupby(['Category', 'Ship Duration']).size().reset_index(name='Order_Count')

fig = px.density_heatmap(
    ship_counts, 
    x="Ship Duration", 
    y="Category", 
    z="Order_Count",
    color_continuous_scale="Viridis",
    title="Frequency Heatmap: How many days does each Category take?",
    labels={'Ship Duration': 'Days to Ship', 'Order_Count': 'Number of Orders'}
)
fig.show()
```

```python
data.info()
```

# Inefficiency analysis through region and category

```python
efficiency_data = data.groupby(['Region', 'Category'])['Profit'].sum().reset_index()

fig = px.density_heatmap(
    efficiency_data, 
    x="Category", 
    y="Region", 
    z="Profit",
    text_auto=True, 
    color_continuous_scale="RdYlGn", 
    title="Regional Inefficiency: Profitability by Category",
    labels={'Profit': 'Total Profit ($)'}
)
fig.show()
```

# Shipping delay analysis

```python
delay_data = data.groupby(['Region', 'Category'])['Ship Duration'].mean().reset_index()

fig_delay = px.bar(
    delay_data, 
    x="Category", 
    y="Ship Duration", 
    color="Category",
    facet_col="Region", 
    title="Operational Inefficiency: Average Shipping Delay per Region",
    labels={'Ship Duration': 'Average Days to Ship'}
)
fig_delay.show()
```

# Segment and category analysis

```python
customer_trend_data = data.groupby(['Segment', 'Category']).size().reset_index(name='Order_Count')

fig = px.density_heatmap(
    customer_trend_data, 
    x="Category", 
    y="Segment", 
    z="Order_Count",
    text_auto=True, 
    color_continuous_scale="Blues",
    title="Customer Shopping Trends: Segment vs. Category",
    labels={'Order_Count': 'Number of Orders'}
)
fig.show()
```

# Preference Distribution

```python
preference_data = data.groupby(['Segment', 'Ship Mode']).size().reset_index(name='Count')

fig = px.bar(
    preference_data, 
    x="Segment", 
    y="Count", 
    color="Ship Mode",
    title="Shipping Preference by Customer Segment (Proportional)",
    labels={'Count': 'Number of Orders'},
    barmode='relative', 
    text_auto=True      
)

fig.update_layout(barnorm='percent') 

fig.show()
```

```python
fig_sun = px.sunburst(
    data, 
    path=['Segment', 'Ship Mode'], 
    values='Quantity', # Or 'Sales' to see which mode generates more revenue
    title="Customer Segment & Shipping Mode Hierarchy",
    color='Segment',
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig_sun.show()
```

# Category and segment analysis through shipping modes

```python
top_orders = data.groupby(['Category', 'Segment', 'Ship Mode']).size().reset_index(name='Order_Count')

fig = px.bar(
    top_orders, 
    x="Category", 
    y="Order_Count", 
    color="Ship Mode",
    facet_col="Segment",
    title="Order Volume: Category vs. Segment vs. Ship Mode",
    labels={'Order_Count': 'Number of Orders'},
    template="plotly_white"
)

fig.show()
```

# Shipping Duration Analysis by Mode

```python
data['Ship Duration'] = (data['Ship Date'] - data['Order Date']).dt.days

fig_ship = px.box(
    data, 
    x="Ship Mode", 
    y="Ship Duration", 
    color="Ship Mode",
    title="Shipping Delay: Which Mode Takes the Longest?",
    labels={'Ship Duration': 'Days to Ship'}
)
fig_ship.show()
```

# Sales Rate & Profit Alteration

```python
time_data = data.resample('ME', on='Order Date')[['Sales', 'Profit']].sum().reset_index()

fig_trend = go.Figure()
fig_trend.add_trace(go.Scatter(x=time_data['Order Date'], y=time_data['Sales'], name='Sales Rate'))
fig_trend.add_trace(go.Scatter(x=time_data['Order Date'], y=time_data['Profit'], name='Profit Margin'))

fig_trend.update_layout(title="Sales Rate vs. Profit Alteration Over Time")
fig_trend.show()
```

# Weekly Trend Analysis

```python
data['Sales'] = data['Sales'].astype(float)

weekly_trend = data.groupby('Order Day of Week')['Sales'].sum().reset_index()

fig_day = px.bar(
    weekly_trend, 
    x='Order Day of Week', 
    y='Sales', 
    title='Total Sales Volume by Day of the Week',
    color='Sales',
    color_continuous_scale='Viridis',
    text_auto='.2s' 
)

fig_day.update_layout(yaxis=dict(range=[0, weekly_trend['Sales'].max() * 1.1]))

fig_day.show()
```

# Monthly Seasonality trends

```python
seasonal_data = data.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
fig_season = px.line(
    seasonal_data, 
    x='Order Month', 
    y='Sales', 
    color='Order Year',
    title='Monthly Seasonality Trends',
    markers=True
)
fig_season.show()
```

# Seasonality : Sales analysis by year and month

```python
heatmap_data = data.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()

pivot_table = heatmap_data.pivot(index='Order Year', columns='Order Month', values='Sales')

fig_heat = px.imshow(
    pivot_table,
    labels=dict(x="Month", y="Year", color="Total Sales"), 
    color_continuous_scale='Viridis',
    title="Seasonality Heatmap: Sales Analysis by Year & Month",
    text_auto='.2s' 
)

fig_heat.show()
```

# Corporate Segment seasonal trends

```python
corporate_data = data[data['Segment'] == 'Corporate']

corporate_seasonal = corporate_data.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()

fig_corp = px.line(
    corporate_seasonal, 
    x='Order Month', 
    y='Sales', 
    color='Order Year',
    title='Corporate Segment: Seasonal Sales Trends',
    markers=True,
    category_orders={"Order Month": ['January', 'February', 'March', 'April', 'May', 'June', 
                                     'July', 'August', 'September', 'October', 'November', 'December']}
)

fig_corp.show()
```

# Discount Analysis

```python
fig_discount = px.scatter(
    data, 
    x="Discount", 
    y="Profit", 
    color="Category",
    hover_data=['Sub-Category', 'Sales'],
    title="The Discount Trap: How Discount Rates Impact Profitability",
    template="plotly_white"
)
fig_discount.show()
```

# Sub-Category "Pareto" Analysis 

```python
sub_category_sales = data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).reset_index()

fig_pareto = px.bar(
    sub_category_sales, 
    x="Sub-Category", 
    y="Sales", 
    color="Sales",
    color_continuous_scale="Viridis",
    title="Pareto Analysis: High-Volume Sub-Categories",
    text_auto='.2s'
)
fig_pareto.show()
```

# Regional Product Preference (Sunburst Chart)

```python
fig_sunburst = px.sunburst(
    data, 
    path=['Region', 'Category', 'Sub-Category'], 
    values='Sales',
    color='Region',
    title="Regional Product Preference Hierarchy",
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig_sunburst.show()
```

# Customer Loyalty (Repeat Purchase Frequency)

```python
loyalty_data = data.groupby('Customer ID').size().reset_index(name='Order_Count')

fig_loyalty = px.histogram(
    loyalty_data, 
    x="Order_Count",
    title="Customer Loyalty Distribution: How Often Do People Return?",
    labels={'Order_Count': 'Number of Orders per Customer', 'count': 'Number of Customers'},
    
)
fig_loyalty.show()
```
